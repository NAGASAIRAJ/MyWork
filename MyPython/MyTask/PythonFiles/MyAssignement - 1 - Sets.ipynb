## Sets

```python
s=()
print(s,'\n')

print(set("Hello World"),'\n')

s=set((1,3,1,4,6,"Hello","Thrymr","Hello"))
print(set([1,3,1,4,6,"Hello","Thrymr","Hello"]),'\n')

s.add(13)
s.add((6,9))
print(s,end=' ')
print('\n')

for i in range(1,6):
    s.add(i)
print(s)
print()

s.update([10,15])
print(s,'\n')

s.remove((6,9))
print(s,'\n')

for i in range(1,5):
    s.discard(i)
print(s,'\n')

s.pop()
s.clear()
print(s,'\n')

String = ('G', 'e', 'e', 'k', 's', 'F', 'o', 'r') 
Fset1 = frozenset(String) 
print(Fset1,'\n') 

s1= {'G', 'e', 'e', 'k', 's', 'F', 'o', 'r'}
s2={'e','s','F'}
s3={'a','b','c'}

s1.add('d')
print(s1,'\n')

s1=s1.copy()
print(s1,'\n')

print(s1.intersection(s2),'\n')

print(s1.isdisjoint(s3),'\n')

print(s2.issubset(s1),'\n')

print(s1.issuperset(s2),'\n')

print(print(s1^s2),' ',print(s1.symmetric_difference(s2)),'\n')

print(print(2*3,end=' '),' ',print(print(2*9,end=' '),' ',print(9//2,end=' ')),'\n')

print(s1.symmetric_difference_update(s2),'\n')
```
